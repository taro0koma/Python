<h3><a name="p2"><span class="sanchor">&nbsp;</span>イベント駆動方式</a></h3>
<p>　GUIアプリケーションはユーザからの入力やシステムの状態の変化などのある事象(イベント：event)をきっかけにプログラムが動きます。このイベントをきっかけにしてプログラムが動くことを<em>イベント駆動・イベントドリブン(eventdriven)</em>と呼びます。イベントドリブン型のプログラムは、一般的に次のようなメインルーチンを持っています。<br>
　1. 初期化<br>
　2. イベントを取得する<br>
　3. イベントの種類に応じて処理を振り分ける<br>
　4. 2～3の繰り返し<br>
</p>
<ul>
<li>2～4を「イベントループ」と呼び、プログラムはユーザからの入力などのイベントを待ちます。</li>
<li>3の処理に対応する機能を「バインディング(binding)」と呼び、イベントが発生したときに、それに応じて定義したプログラムを実行します。このプログラムを「イベントハンドラ」とか「コールバック」と呼びます。</li>
</ul>

<h2><a name="p10">フォント指定</a></h2>
<p>個々のウィジェットのフォントはオブジェクト生成時のオプション(font)で指定することができます。しかし、全てのラベル・ウィジェットで共通に使用するフォントを指定したい場合もあります。この場合はoption_add()メソッドを使ってデフォルト値を設定します。使い方は次のとおりです。<br>
</p>
<pre>&gt;&gt;&gt; import tkinter as tk

&gt;&gt;&gt; root = tk.Tk()

&gt;&gt;&gt; root.geometry(&quot;200x150&quot;)

&gt;&gt;&gt; root.option_add('*font', 'FixSys 14')

&gt;&gt;&gt; tk.Label(text = 'Hello').pack()

&gt;&gt;&gt; tk.Label(text = 'Python').pack()
</pre>
<br>
<p>root はメインウィンドウを表します。これで、テキストを表示するウィジェットは、指定したフォントを使って表示されます。第 1 引数はデフォルト値を設定するウィジェットを表すパターンです。第 2 引数が設定する値です。第 3 引数は省略していますが、優先順位 priority (0 - 100) を設定することができます。<br>
</p>
<p>パターンは、アプリケーション名、ウィジェット名、オプション名をドットで区切って表しますが、ワイルドカード * やウィジェットを表すクラス名を指定することもできます。*font の場合は、アプリケーションで使用するフォントを指定することになります。ラベルに対してフォントを設定したい場合は *Label.font となります。<br>
</p>

<h3><a name="p12"><span class="sanchor">&nbsp;</span> packメソッド </a></h3>
<p>pack は widget を縦または横に１次元的に配列するときに使います。<br>
プログラム例<br>
</p>

```python
import tkinter as tk
 
root = tk.Tk()
root.title("packer")
root.geometry("200x150")
color = ('red', 'blue', 'green', 'yellow')
 
#ラベルの作成
for x in range(4):
    label = tk.Label(root, text = "Label{}".format(x) , background = color[x])
    label.pack()
 
root.mainloop()
 
```

<h3><a name="p14"><span class="sanchor">&nbsp;</span> gridメソッド </a></h3>
<p>２次元的に配置するときは grid を使います。<br>
詳細は以下のURLを参照してください。<br>
<a href="http://infohost.nmt.edu/tcc/help/pubs/tkinter/web/grid.html" target="_blank">http://infohost.nmt.edu/tcc/help/pubs/tkinter/web/grid.html</a><br>

```python
import tkinter as tk
 
root = tk.Tk()
root.title("gridder")
root.geometry("200x150")
color = ('red', 'blue', 'green', 'yellow')
 
#ラベルの作成
label1 = tk.Label(root, text = "Label1", background = color[0])
label2 = tk.Label(root, text = "Label2", background = color[1])
label3 = tk.Label(root, text = "Label3", background = color[2])
label4 = tk.Label(root, text = "Label4", background = color[3])
 
label1.grid(column=0,row=0)
label2.grid(column=1,row=0)
label3.grid(column=0,row=1)
label4.grid(column=1,row=1)
 
root.mainloop()
```

<h3><a name="p15"><span class="sanchor">&nbsp;</span>placeメソッド</a></h3>
<p>配置を直接指定する方法です。pack や grid でうまく配置できないときに使います。place の option は以下の通りです。<br>
<div class="wikitable"  style="">
<table style="">
<tr><th colspan='2'></th></tr>
<tr><td>anchor	</td><td>pack と同じ、デフォルトは Tk.NW</td></tr>
<tr><td>bordermode	</td><td>縁を内側 (Tk.INSIDE) か外側(Tk.OUTSIDE)につける。デフォルトは Tk.INSIDE</td></tr>
<tr><td>height	</td><td>高さを pixel で指定。</td></tr>
<tr><td>relheight	</td><td>親 widget に対する相対的な高さを 0.0 -- 1.0 の実数で指定。</td></tr>
<tr><td>relwidth	</td><td>親 widget に対する相対的な幅を 0.0 -- 1.0 の実数で指定。</td></tr>
<tr><td>relx	</td><td>親 widget に対する相対的な横の位置を 0.0 -- 1.0 の実数で指定。デフォルトは 0</td></tr>
<tr><td>rely	</td><td>親 widget に対する相対的な横の位置を 0.0 -- 1.0 の実数で指定。デフォルトは 0</td></tr>
<tr><td>width	</td><td>幅を pixel で指定。</td></tr>
<tr><td>x	</td><td>横の位置を pixel で指定。デフォルトは 0</td></tr>
<tr><td>y	</td><td>縦の位置を pixel で指定。デフォルトは 0</td></tr>
</table>
</div>
</p>
<ul>
<li>placeメソッドのプログラム例(オプション：relx  rely を使用）</li>
</ul>

```python
import tkinter as tk
 
root = tk.Tk()
root.title("placer")
root.geometry("200x160")
 
color = ("red","blue","green")
for x in range(3):
    lab = tk.Label(root, text = "Label{}".format(x), bg = color[x])
    lab.place( relx=0.25*x, rely=0.25*x)
 
root.mainloop()
```

```python
# placeメソッドのプログラム例(オプション：x y を使用）
import tkinter as tk
 
root = tk.Tk()
root.title("placer")
root.geometry("200x160")
 
color = ("red","blue","green")
for x in range(3):
    lab = tk.Label(root, text = "Label{}".format(x), bg = color[x])
    lab.place( x = 50*x, y = 40*x)
 
root.mainloop()
```

<h2><a name="p16">ウィジェット変数を関連付ける </a></h2>
<p><a href="https://docs.python.jp/3/library/tkinter.html" target="_blank">https://docs.python.jp/3/library/tkinter.html</a>　より抜粋<br>
</p>
<p>ウィジェットによっては、(テキスト入力ウィジェットのように) 特殊なオプションを使って、現在設定されている値をアプリケーション内の変数に直接関連付けできます。<br>
次のオプションがそれに該当します。<br>
</p>
<ul>
<li> variable</li>
<li> textvariable</li>
<li> onvalue</li>
<li> offvalue</li>
<li> value </li>
</ul>
<p>この関連付けは双方向に働きます: 変数の値が何らかの理由で変更されると、関連付けされているウィジェットも更新され、新しい値を反映します。<br>
</p>
<p>残念ながら、現在の tkinter の実装では、<em>variable</em> や <em>textvariable</em> オプションでは任意の Python の値をウィジェットに渡せません。この関連付け機能がうまく働くのは、tkinter 内で Variable というクラスからサブクラス化されている変数によるオプションだけです。<br>
</p>
<p>Variable には、 <em>StringVar, IntVar, DoubleVar, BooleanVar</em> といった便利なサブクラスがすでに数多く定義されています。こうした変数の現在の値を読み出したければ、 get() メソッドを呼び出します。また、値を変更したければ set() メソッドを呼び出します。このプロトコルに従っている限り、それ以上なにも手を加えなくてもウィジェットは常に現在値に追従します。<br>
</p>
<ul>
<li>実行例</li>
</ul>
<pre>&gt;&gt;&gt; import tkinter as tk

&gt;&gt;&gt; root = tk.Tk()

&gt;&gt;&gt; root.geometry(&quot;200x150&quot;)

&gt;&gt;&gt; lab_str = tk.StringVar()

&gt;&gt;&gt; lab_str.set('Hello')

&gt;&gt;&gt; lab = tk.Label(textvariable=lab_str, bg = 'yellow')

&gt;&gt;&gt; lab.pack()

&gt;&gt;&gt; lab_str.set('Python')
</pre>
